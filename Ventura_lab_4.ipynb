```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

A = np.array([[-0.4], [4.3], [-0.6]])
B = np.array([-0.2, 0.21, 0])
C = np.array([-0.22, 1, -1.5])

print("Vector A:", "\n", A)
print("Vector B:", "\n", B)
print("Vector C:", "\n", C, "\n")

add = A + B
dot = np.dot(C, A)
cross = np.cross(B, C)

print("A + B:", "\n", add)
print("Dot Product of C and A:", "\n", dot)
print("Cross Product of B and C:", "\n", cross)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(0, 0, 0, cross[0], cross[1], cross[2], color='r', label ='add')
ax.quiver(0, 0, 0, dot[0], dot[0], dot[0], color='g', label='dot')
ax.quiver(0, 0, 0, add[0], add[1], add[2], color='b', label='cross')
ax.set_xlim([-6, 6])
ax.set_ylim([-6, 6])
ax.set_zlim([-6, 6])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()
plt.show()
```
